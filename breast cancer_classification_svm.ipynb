```python
#splitting data into features and target for model building
```

```python
x=can_df.drop(['target'],axis=1)
x.head()
```

```python
y=can_df['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
#model predicting
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```

```python
#normalizing dataset
```

```python
train_min=x_train.min()
train_min
```

```python
train_range=x_train.max()-x_train.min()
train_range
```

```python
x_train_scaled = (x_train-train_min)/train_range
x_train_scaled
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x=x_train['mean area'],y=x_train['mean smoothness'],hue=y_train)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x=x_train_scaled['mean area'],y=x_train_scaled['mean smoothness'],hue=y_train)
plt.show()
```

```python
#normalizing test set
```

```python
min_test=x_test.min()
min_test
```

```python
test_range=x_test.max()-x_test.min()
test_range
```

```python
x_test_scaled = (x_test-min_test)/test_range
x_test_scaled
```

```python
#building new model
```

```python
svc.fit(x_train_scaled,y_train)
```

```python
y_pred2=svc.predict(x_test_scaled)
```

```python
y_pred2
```

```python
cm2=confusion_matrix(y_test,y_pred2)
cm2
```

```python
sns.heatmap(cm2,annot=True)
plt.show()
```

```python
#implementing grid search
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
grid
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(x_test_scaled)
grid_pred
```

```python
cm=confusion_matrix(y_test,grid_pred)
cm
```
